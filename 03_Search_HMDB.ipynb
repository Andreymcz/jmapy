```python
import config
import cutils as utils
import hmdb_web_db as hmdb_web
import pandas as pd

folder = "D:\\NAS\\TEST\\"
file = folder + "DATA.xlsx"
export_file = folder + "DATA_output.xlsx"
column_name = "HMDB ID"
info_titles = ["Common Name"]


export_columns = [column_name]
columns = export_columns + info_titles
output_data = {title : [] for title in columns}

compounds_table = pd.read_excel(file)

#####
num_rows = compounds_table.shape[0]
row_count = 0
for row_index, row_data in compounds_table.iterrows() :
    hmdb_ids = row_data.at[column_name]
    print('({:d}/{:d})Searching HMDB info: '.format(row_count, num_rows), hmdb_ids)
    row_count+=1
    
    #### search hmdb info
    hmdb_info = None
    if not pd.isna(hmdb_ids) and len(str(hmdb_ids)) > 0 and type(hmdb_ids) is str:
        for hmdb_id in hmdb_ids.split(';'):
            hmdb_info = hmdb_web.search_metabolite_info_by_id(hmdb_id, info_titles)            
            break
    
    #### set hmdb data
    if hmdb_info != None:
        for column_name, value in hmdb_info.items():        
            output_data[column_name].append(value)
    else:
        for column_name in info_titles:
            output_data[column_name].append("")
        
    #### set compounds data
    for column_name in export_columns:
        output_data[column_name].append(row_data[column_name])
        
            
result = pd.DataFrame.from_dict(output_data)
result.to_excel(export_file)
display(result)
```
