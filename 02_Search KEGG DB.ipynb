```python
import config
import cutils as utils
import keeg_web_db as keeg_web
import pandas as pd

folder = "D:\\NAS\\TEST\\"
keeg_ids_file = folder + "Keeg_id.xlsx"
keeg_id_column_name = "KEGG ID"
export_file = folder + "Keeg_info_output.xlsx"



info_titles = ['PubChem', 'ChEBI', 'ChEMBL', 'Name']
compounds_columns = [keeg_id_column_name]

columns = compounds_columns + info_titles
output_data = {title : [] for title in columns}

compounds_table = pd.read_excel(keeg_ids_file)

num_rows = compounds_table.shape[0]
row_count = 0
for row_index, row_data in compounds_table.iterrows() :
    print(row_data)
    keeg_ids = row_data.at[keeg_id_column_name]
    print('({:d}/{:d})Searching Keeg info: '.format(row_count, num_rows), keeg_ids)
    row_count+=1
    
    #### search keeg info
    keeg_info = None
    if not pd.isna(keeg_ids) and len(str(keeg_ids)) > 0 and type(keeg_ids) is str:
        for keeg_id in keeg_ids.split(';'):
            keeg_info = keeg_web.search_keeg_info_by_id(keeg_id, info_titles)
            print(keeg_info)
            break
    
    #### set keeg data
    if keeg_info != None:
        for column_name, value in keeg_info.items():        
            output_data[column_name].append(value)
    else:
        for column_name in info_titles:
            output_data[column_name].append("")
        
    #### set compounds data
    for column_name in compounds_columns:
        output_data[column_name].append(row_data[column_name])
        

result = pd.DataFrame.from_dict(output_data)
display(result)
result.to_excel(export_file)

#keeg_web.search_keeg_info_by_id('C10129', ['PubChem', 'ChEBI', 'ChEMBL', 'Name'])
```
